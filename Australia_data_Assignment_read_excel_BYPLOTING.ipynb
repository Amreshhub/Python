```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
```

```python
df= pd.read_excel('Australia_Filtered.xlsx')
df
```

#Plot bar chart of total no of Emigrants for countires top 5 countires having max no of total emigrants



```python
a = df.sort_values(by ='total').tail()
plt.bar(a.Country.values, a.total.values)
plt.xticks(rotation = 90)

a.plot(x = 'Country', y = 'total', kind = 'bar')
plt.show()

```

```python
#Find total Emigrants from each category by area name and plot barchart for each area
grp = df.groupby('AreaName')
d = grp.total.sum()
print(d)

d.plot(kind='bar')
plt.show()
```

```python
plt.bar(d.index,d.values)

plt.xticks(rotation='90')
plt.show()
```

```python
#Plot total Emigration trend(emigration of world over the years i.e. year vs Total count for that year) using line plot
w = df.iloc[:, 5:-1].sum()
print(w.to_frame().T)

w.plot(kind = 'line')
```

```python
plt.plot(w.index,w.values)
plt.show()
```

```python
#Compare Emigration Trend for Asia, Africa and Europe AreaName (Use Line plot).
plt.figure(figsize = (16,6))
grp = df.groupby('AreaName')

asia = grp.get_group('Asia').iloc[:, 5:-1].sum()
plt.plot(asia, label='Asia')

asia = grp.get_group('Africa').iloc[:, 5:-1].sum()
plt.plot(asia, label='Africa')

asia = grp.get_group('Europe').iloc[:, 5:-1].sum()
plt.plot(asia, label='Europe')

plt.legend()
plt.xticks(rotation= 90)
plt.show()
```

```python
#Plot Total Emmigrants based on DevelopmenRegion
plt.figure(figsize=(16,6))
years = map(str, list(range(1980,2009)))
years = list(years)
print(years)


r = df.groupby('AreaName')[years].sum()
for row in r.index:
    plt.plot(r.loc[row],label = row)
    
    
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
#Using matlplot
plt.figure(figsize = (16,6))
df1 = df.groupby('DevName').total.sum()
plt.barh(df1.index, df1.values, height = 0.7, )
plt.ylabel('Region')
plt.xlabel('Immigrants')
plt.show()
```

```python
#Pandas
df1.plot(kind= 'barh', figsize = (16,6))
plt.xlabel('Emigrants')
plt.legend()
plt.show()
```

```python
#Plot Emigration trend for India using line plot
```

```python
idx = df.index[df.Country == 'India']
data = df.loc[idx].iloc[:, 5: -1].T
data.plot(kind = 'line')
plt.show()
```

```python
#Plot Emigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot
data= df.sort_values(by='total',ascending= False).head()
label = data.Country
values = data.iloc[:, 5:-1]

df1 = pd.DataFrame(values.T)
df1.columns = label


df1.plot(kind = 'line')
plt.legend(fontsize=5)
plt.show()
```

```python
#- For the above use area plot with alpha = 0.5
plt.figure(figsize =(16,6))
data = df.sort_values(by = 'total',ascending= False).head()
label = data.Country
values = data.iloc[:,5: -1]

#dataFram
df1 = pd.DataFrame(values.T)
df1.columns = label

df1.plot(kind ='area', stacked = False, alpha = 0.5)
plt.legend(fontsize=5)
plt.show()
```

```python
# Compare Immigration Trend for top-5 countires using stacked area plot
# For the above use area plot with alpha = 0.5
plt.figure(figsize=(16,5))
data = df.sort_values(by = 'total', ascending= False).head()
label = data.Country
values = data.iloc[:, 5: -1]


df1 = pd.DataFrame(values.T)
df1.columns = label

df1.plot(kind = 'area', stacked = True)
plt.legend(fontsize=5)
plt.show()
```
